```python
import pandas as pd

prepared_path = r"C:\Users\User\Downloads\termekleirasok_elso_20.csv"
prepared_21_40_path = r"C:\Users\User\Downloads\termeklista_21-40_javitott.csv"
prepared_41_60_path = r"C:\Users\User\Downloads\termeklista_41-60.csv"
target_path = r"C:\Users\User\Downloads\UnasShop_toys4all-unas-hu_20250903-162158.csv"

# Beolvasás
prepared = pd.read_csv(
    prepared_path,
    sep=",",
    encoding="utf-8",
    quotechar='"',
    engine="python",
    dtype={"Cikkszám": "string"}
)

prepared_21_40 = pd.read_csv(
    prepared_21_40_path,
    sep=",",
    encoding="utf-8",
    dtype=str
)

prepared_41_60 = pd.read_csv(
    prepared_41_60_path,
    sep=",",
    encoding="utf-8",
    dtype=str
)

target = pd.read_csv(
    target_path,
    sep=None,
    engine="python",
    dtype=str,
    encoding="utf-8",
)

# Oszlopnevek normalizálása (BOM, idézőjelek, szóközök körben)
normalize = lambda s: s.replace('\ufeff', '').strip().strip('"')
prepared.columns = [normalize(c) for c in prepared.columns]
prepared_21_40.columns = [normalize(c) for c in prepared_21_40.columns]
prepared_41_60.columns = [normalize(c) for c in prepared_41_60.columns]
target.columns = [normalize(c) for c in target.columns]

# Az első 20 sor közös oszlopainak cseréje
n = min(20, len(prepared), len(target))
common_cols = [c for c in prepared.columns if c in target.columns]
new_df = target.copy()
if n > 0 and common_cols:
    new_df.loc[new_df.index[:n], common_cols] = prepared.loc[prepared.index[:n], common_cols].values

# A 21-40. termékek hozzáadása (20-39. indexek)
n_21_40 = min(20, len(prepared_21_40), len(target) - 20)
if n_21_40 > 0 and common_cols:
    start_idx = 20
    end_idx = start_idx + n_21_40
    new_df.loc[new_df.index[start_idx:end_idx], common_cols] = prepared_21_40.loc[prepared_21_40.index[:n_21_40], common_cols].values

# A 41-60. termékek hozzáadása (40-59. indexek)
n_41_60 = min(20, len(prepared_41_60), len(target) - 40)
if n_41_60 > 0 and common_cols:
    start_idx = 40
    end_idx = start_idx + n_41_60
    new_df.loc[new_df.index[start_idx:end_idx], common_cols] = prepared_41_60.loc[prepared_41_60.index[:n_41_60], common_cols].values

# Biztosítsuk, hogy a 'Cikkszám' benne legyen és első legyen
if "Cikkszám" not in new_df.columns and "Cikkszám" in target.columns:
    new_df.insert(0, "Cikkszám", target["Cikkszám"])

# A 'Tömeg' oszlop maradjon a cél CSV-ből (ha létezik)
if "Tömeg" in target.columns:
    new_df["Tömeg"] = target["Tömeg"]

# Csak a kért oszlopok maradjanak, ebben a sorrendben
selected_cols = [
    "Cikkszám",
    "Termék Név",
    "Nettó Ár",
    "Bruttó Ár",
    "Rövid Leírás",
    "Tulajdonságok",
]
cols_present = [c for c in selected_cols if c in new_df.columns]
# Ha a Cikkszám hiányzott a listából, de most már benne van a DF-ben, tegyük az élére
if "Cikkszám" in new_df.columns and "Cikkszám" not in cols_present:
    cols_present = ["Cikkszám"] + cols_present
new_df = new_df.loc[:, cols_present]

# Eredmény előnézete (első 25 sor)
new_df.iloc[:25]
```

```python
# Hiányos sorok számlálása (nincs Rövid Leírás ÉS nincs Tulajdonságok)
missing_both = new_df[
    (new_df["Rövid Leírás"].isna() | (new_df["Rövid Leírás"].str.strip() == "")) &
    (new_df["Tulajdonságok"].isna() | (new_df["Tulajdonságok"].str.strip() == ""))
]

print(f"Összes sor: {len(new_df)}")
print(f"Hiányos sorok (nincs Rövid Leírás ÉS Tulajdonságok): {len(missing_both)}")
print(f"Teljes sorok: {len(new_df) - len(missing_both)}")

# Ha szeretnéd látni a hiányos sorok cikkszámát:
if len(missing_both) > 0:
    print("\nHiányos sorok cikkszámai:")
    print(missing_both["Cikkszám"].tolist())
```

```python
# 41-60. termékek statisztikája
print(f"41-60. termékek fájlban:")
print(f"- Összes sor: {len(prepared_41_60)}")
print(f"- Cikkszámok száma: {len(prepared_41_60['Cikkszám'].unique())}")

# Teljes adatokkal rendelkező sorok (van Rövid Leírás ÉS Tulajdonságok)
complete_41_60 = prepared_41_60[
    (~prepared_41_60["Rövid Leírás"].isna()) & (prepared_41_60["Rövid Leírás"].str.strip() != "") &
    (~prepared_41_60["Tulajdonságok"].isna()) & (prepared_41_60["Tulajdonságok"].str.strip() != "")
]

print(f"- Teljes adatokkal rendelkező sorok: {len(complete_41_60)}")
print(f"- Hiányos sorok: {len(prepared_41_60) - len(complete_41_60)}")

# Cikkszámok listája
print(f"\n41-60. termékek cikkszámai:")
print(prepared_41_60["Cikkszám"].tolist())
```
